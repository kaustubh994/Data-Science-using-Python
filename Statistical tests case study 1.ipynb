```python
#4 Is there any difference between segments of customers in terms of credit card usage ?
```

```python
p.groupby("segment")["pre_usage"].mean()
```

```python
stats.ttest_ind(p["segment"],p["pre_usage"])
```

```python
#It shows the comparision between the means of the card usage of the different segment of customers. Also, since the p-value is
# less than 0.05 which signifies that there is some difference between segments of customers in terms of credit card usage
```

```python
#5 Is there any relation between region & Segment ?
```

```python
p["region"].corr(p["segment"])
```

```python
stats.ttest_ind(p["region"],p["segment"])
```

```python
# The above correlation shows the weak linear relationship between the customers in the region and segment
```

```python
#6 Is the relationship between Card usage in the latest month and pre usage of campaign ?
```

```python
p["pre_usage"].corr(p["Latest_mon_usage"])
```

```python
stats.ttest_ind(p["Latest_mon_usage"],p["pre_usage"])
```

```python
# The above correlation shows the positive relationship between the pre-usage and latest-month-usage customers. Also, since the 
# p-value is less than 0.05 which signifies that there is some relationship between Card usage in the latest month and pre 
# usage of campaign
```
