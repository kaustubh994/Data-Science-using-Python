### Step 10.  What is the difference in days between the first day and the oldest

```python
d0 = date(1980, 12, 11)
d1 = date(2014, 7, 7)
delta = d1 - d0
```

```python
delta.days
```

### Step 11.  How many months in the data we have?

```python
dates = [dt for dt in rrule(MONTHLY, dtstart=d0, until=d1)]
```

```python
len(dates)
```
