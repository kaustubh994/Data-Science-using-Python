```python
## Here, patients with the age group of more than 65 are more unique than others
##       patients with the income more than 16000 are more unique than others
##       Males are less unique than the females
```

```python
def incometyp(patient):
    if (patient['income']  == '16000-23999'):
        return 2
    elif (patient['income']  == '24000-31999'):
        return 3
    elif (patient['income']  == '32000-47999'):
        return 4
    elif (patient['income']  == '48000+'):
        return 5
    else:
        return 1
```

```python
def agetyp(patient):
    if (patient['age']  == '65-74'):
        return 2
    elif (patient['age']  == '75-84'):
        return 3
    elif (patient['age']  == '85+'):
        return 4
    else:
        return 1
```

```python
def gendertyp(patient):
    if (patient['gender']  == 'M'):
        return 1
    else:
        return 0
```

```python
r8["Income group"]=r8.apply(incometyp,axis=1)
```

```python
r8["Age group"]=r8.apply(agetyp,axis=1)
```

```python
r8["Gender group"]=r8.apply(gendertyp,axis=1)
```

```python
r8.drop(['age', 'gender', 'income'], axis =1 , inplace = True)
```

```python
r9=r8[["Income group","Age group","Gender group"]]
r9
```
