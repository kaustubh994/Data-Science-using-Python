```python
## Building Models
```

```python
## Firstly we will set up the data by dividing it into training, validation and testing set according to the start year of the
## base table
```

```python
basetable_new["Start year"].min()
```

```python
basetable_new["Start year"].max()
```

```python
train=basetable_new[basetable_new["Start year"]<2001]
```

```python
validation=basetable_new[(basetable_new["Start year"]>=2001) & (basetable_new["Start year"]<2003)]
```

```python
test=basetable_new[basetable_new["Start year"]>=2003]
```

```python
x_train = train.drop(['Churn'], axis=1)
```

```python
y_train = train["Churn"]
```

```python
x_validation = validation.drop(['Churn'], axis=1)
```

```python
y_validation = validation["Churn"]
```

```python
x_test = test.drop(['Churn'], axis=1)
```

```python
y_test = test["Churn"]
```

```python
# Logistic Regression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr_train=accuracy_score(lr.predict(x_train), y_train)*100
lr_train
```

```python
lr_validation=accuracy_score(lr.predict(x_validation), y_validation)*100
lr_validation
```

```python
lr_test=accuracy_score(lr.predict(x_test), y_test)*100
lr_test
```

```python
# Support Vector Machine (SVM)
```

```python
svm=SVC()
```

```python
svm.fit(x_train, y_train)
```

```python
svm_train=accuracy_score(svm.predict(x_train), y_train)*100
svm_train
```

```python
svm_validation=accuracy_score(svm.predict(x_validation), y_validation)*100
svm_validation
```

```python
svm_test=accuracy_score(svm.predict(x_test), y_test)*100
svm_test
```

```python
# K-nearest Neighbours (KNN)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
knn_train=accuracy_score(knn.predict(x_train), y_train)*100
knn_train
```

```python
knn_validation=accuracy_score(knn.predict(x_validation), y_validation)*100
knn_validation
```

```python
knn_test=accuracy_score(knn.predict(x_test), y_test)*100
knn_test
```
